```python
main_list = [21, 22, 23, 24, 25]
sub_list = [22, 23]

sublists = []

for i in range(len(main_list) - len(sub_list) + 1):
    temp = main_list[i:i+len(sub_list)].copy()
    sublists.append(temp)

if sub_list in sublists:
    print("Sublist Have")
else:
    print("Sublist Haven't")

```
